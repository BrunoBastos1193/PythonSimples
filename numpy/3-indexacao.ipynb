```python
import numpy as np
```

```python
array = np.array([ [10, 20 , 30],
                   [40, 50, 60],
                   [70, 80, 90]])

print(array.ndim)
print(array.shape)
array
```

```python
#Buscar o elemento na linha 1 , coluna 2
elemento = array[1, 2]
elemento
```

```python
#Pegar os dados da primeira linha
linha_0 = array[0]
linha_0
```

```python
#acessando a coluna 1
coluna_1 = array[:, 1]
coluna_1
```

```python
#Fatiando para obter um subarray
#Linha 0 a 1 | Coluna 1 a 2
subarray = array [0:2 , 1:3]
subarray
```

```python
mascara = array > 50
mascara
```

```python
elemento_maior_que_50 = array[mascara]
elemento_maior_que_50
```

```python
#Modificando elemento linha 1 | Coluna 1
array[1, 1] = 100
array
```
