```python
import numpy as np
```

```python
array = np.array([1, 2, 3, 4, 5])
array
```

```python
#1- Função Adição (add)
soma = np.add(array, 10)
soma
```

```python
#2- Função Subtração (Subtract)
subtracao = np.subtract(array, 5)
subtracao
```

```python
#3- Função Multiplicação (Multiply)
produto = np.multiply(array, 2)
produto
```

```python
#4- Função Divisão (Divide)
divisao = np.divide(array, 2)
divisao
```

```python
#5- Função Trigonométricas (Sin) SENO
seno = np.sin(array)
seno
```

```python
#5- Função Trigonométricas (Cos) COSSENO
cosseno = np.cos(array)
cosseno
```

```python
#6- Funções Estatísticas (Mean)
media = np.mean(array)
media
```

```python
#6- Funções Estatísticas (Std)
desvio_padrao = np.std(array)
desvio_padrao
```

```python
#7- Funções de Comparação (<,>,<=,>=)
comparacao_maior_que_3 = array > 3
comparacao_maior_que_3
```

```python
#8- Combinando Funções (Sqrt) Raiz Quadrada + (Power) 
resultado = np.sqrt(array) + np.power(array, 2)
resultado
```

```python
#9- Função Exponencial (Exp)
exponencial = np.exp(array)
exponencial
```

```python
#10- Função de Logaritmo (Log)
logaritmo = np.log(array)
logaritmo
```

```python
#11- Função de arredondamento
arredondado = np.round(array / 2)
arredondado
```

```python
#12- Função de Potência (Power)
potencia = np.power(array, 3)
potencia
```
