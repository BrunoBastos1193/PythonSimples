```python
#Fatiando as duas primeiras linhas
fatiando_linhas = array[0:2, :]
fatiando_linhas
```

```python
#Fatiando as trÃªs primeiras colunas
fatiando_colunas = array [:, 0:3]
fatiando_colunas
```

```python
subarray_central = array [1:3, 1:3]
subarray_central
```

```python
subarray_lateral = array [2:4 , 3:5]
subarray_lateral
```
